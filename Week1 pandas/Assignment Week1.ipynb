# 8. Find all instances where the weather is clear and the relative humidity is greater than 50, or visibility is above 40

```python
clear_relative = (df['Weather_Condition'] == 'Clear') & (df['Rel Hum_%'] > 50)
clear_relative
```

```python
visibility = df['Visibility_km'] > 40
visibility
```

```python
instances = df[clear_relative | visibility]
instances
```
